```python
from sklearn.model_selection import KFold
from sklearn.metrics import mean_absolute_error
from sklearn.ensemble import RandomForestRegressor
```

# RF KFOLD BASE A

```python
#faz a separação 90 / 10
kf= KFold(n_splits=10)
kf.get_n_splits(xA)

todos_mae=[]
for train_index, test_index in kf.split(xA):
    x_train, x_test = xA.values[train_index], xA.values[test_index] #gera os indices a serem treinados
    y_train, y_test = yA.values[train_index], yA.values[test_index] 
    
    
    regr=RandomForestRegressor(criterion='mse')
    regr.fit(x_train,y_train)
    
    previsoes= regr.predict(x_test)
    erro_médio = mean_absolute_error(y_test, previsoes)

    todos_mae.append(erro_médio)

mae_finalA=0
for n in todos_mae:
    mae_finalA+= (n/ len(todos_mae))
```

# RF KFOLD BASE B

```python

#faz a separação 90 / 10
kf= KFold(n_splits=10)
kf.get_n_splits(xB)

todos_mae=[]
for train_index, test_index in kf.split(xB):
    x_train, x_test = xB.values[train_index], xB.values[test_index] #gera os indices a serem treinados
    y_train, y_test = yB.values[train_index], yB.values[test_index] 
    
    
    regr=RandomForestRegressor(criterion='mse')
    regr.fit(x_train,y_train)
    
    previsoes= regr.predict(x_test)
    erro_médio = mean_absolute_error(y_test, previsoes)

    todos_mae.append(erro_médio)

mae_finalB=0
for n in todos_mae:
    mae_finalB+= (n/ len(todos_mae))
```
