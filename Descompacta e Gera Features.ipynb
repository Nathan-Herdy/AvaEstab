# Função que irá adicionar as features e irá salvar o dataframe em features

```python
def statistical_feature (data):
    
    freq = 100
    statistical_features=np.zeros((17,1))

    #Mean value
    mean_value=np.mean(data)
    statistical_features[0]=mean_value

    #Varience
    variance=np.var(data)
    statistical_features[1]=variance

    #Entropy
    from scipy.stats import entropy
    Entropy=entropy(data)
    statistical_features[2]=Entropy

    #Kurtosis
    kur=0
    for x in range(len(data)):
        kur=kur+(((data[x]-mean_value))/(variance**1/2))**4
    
    Kurtosis=kur/len(data)
    statistical_features[3]=Kurtosis

    #Skewness 
    skew=0
    for x in range(len(data)):
        skew=skew+(((data[x]-mean_value))/(variance**1/2))**3
    
    skewness=skew/len(data)
    statistical_features[4]=skewness

    #Root mean square value
    rms=0
    for x in range(len(data)):
        rms=rms+(data[x])**2
    
    root_mean_square_value=(rms/len(data))**1/2
    statistical_features[5]=root_mean_square_value

    #Peak-to-peak value
    ppv=max(data)-min(data)
    statistical_features[6]=ppv

    #Crest factor
    cf=max(data)/root_mean_square_value
    statistical_features[7]=cf

    #Impulse factor
    If=0
    for x in range(len(data)):
        If=If+(data[x]/len(data))
    
    impulse_factor=max(data)/If
    statistical_features[8]=impulse_factor

    #Kurtosis factor
    kurtosis_factor=Kurtosis/(root_mean_square_value)**4
    statistical_features[9]=kurtosis_factor
    
    #Mean Speed
    mvel = np.sum(np.abs(np.diff(data, axis=0)), axis=0) / 30
    statistical_features[10]=mvel
    
    
    #Mean Resultant Speed
    mvelr = np.sum(np.abs(np.sqrt(np.sum(np.diff(data) ** 2))), axis=0) / 30
    statistical_features[11]=mvelr
    
    #frequency percentiles of the power spectral density - 50% 
    fp_ap, mf_ap, fmax_ap, Ptot_ap, F, P_ap = psd(data, fs=freq, show=False)
    statistical_features[12]=fp_ap[50]
    
    #frequency percentiles of the power spectral density - 95%
    statistical_features[13]=fp_ap[95]
    
    #Peaky Frequency
    statistical_features[14]=fmax_ap
    
    #Mean Frequence
    statistical_features[15] = mf_ap
    
    #Total power in units squared.
    statistical_features[16]=pf = Ptot_ap
    
    return statistical_features
```

```python
colzero = len(bds.columns)

def cria_feature(dataframe):
    # cria as features de cop x
    dataframe.insert(len(dataframe.columns), 'x_mean_value_detrended_cop',value=float)
    dataframe.insert(len(dataframe.columns), 'x_variance_cop',value=float)
    dataframe.insert(len(dataframe.columns), 'x_entropy_cop',value=float)
    dataframe.insert(len(dataframe.columns), 'x_kurtosis_cop',value=float)
    dataframe.insert(len(dataframe.columns), 'x_skewness_cop',value=float)
    dataframe.insert(len(dataframe.columns), 'x_rms_cop',value=float)
    dataframe.insert(len(dataframe.columns), 'x_peaky_to_peaky_cop',value=float)
    dataframe.insert(len(dataframe.columns), 'x_crest_factor_cop',value=float)
    dataframe.insert(len(dataframe.columns), 'x_impulse_factor_cop',value=float)
    dataframe.insert(len(dataframe.columns), 'x_kurtosis_factor_cop',value=float)
    dataframe.insert(len(dataframe.columns), 'x_mean_speed_cop',value=float)
    dataframe.insert(len(dataframe.columns), 'x_Mean_Resultant_Speed_cop',value=float)
    dataframe.insert(len(dataframe.columns), 'x_fpcntile50_cop',value=float)
    dataframe.insert(len(dataframe.columns), 'x_fpcntile95_cop',value=float)
    dataframe.insert(len(dataframe.columns), 'x_Peaky_Frequency_cop',value=float)
    dataframe.insert(len(dataframe.columns), 'x_Mean_Frequency_cop',value=float)
    dataframe.insert(len(dataframe.columns), 'x_Total_power_cop',value=float)

    # cria as features de cop y
    dataframe.insert(len(dataframe.columns), 'y_mean_value_detrended_cop',value=float)
    dataframe.insert(len(dataframe.columns), 'y_variance_cop',value=float)
    dataframe.insert(len(dataframe.columns), 'y_entropy_cop',value=float)
    dataframe.insert(len(dataframe.columns), 'y_kurtosis_cop',value=float)
    dataframe.insert(len(dataframe.columns), 'y_skewness_cop',value=float)
    dataframe.insert(len(dataframe.columns), 'y_rms_cop',value=float)
    dataframe.insert(len(dataframe.columns), 'y_peaky_to_peaky_cop',value=float)
    dataframe.insert(len(dataframe.columns), 'y_crest_factor_cop',value=float)
    dataframe.insert(len(dataframe.columns), 'y_impulse_factor_cop',value=float)
    dataframe.insert(len(dataframe.columns), 'y_kurtosis_factor_cop',value=float)
    dataframe.insert(len(dataframe.columns), 'y_mean_speed_cop',value=float)
    dataframe.insert(len(dataframe.columns), 'y_Mean_Resultant_Speed_cop',value=float)
    dataframe.insert(len(dataframe.columns), 'y_fpcntile50_cop',value=float)
    dataframe.insert(len(dataframe.columns), 'y_fpcntile95_cop',value=float)
    dataframe.insert(len(dataframe.columns), 'y_Peaky_Frequency_cop',value=float)
    dataframe.insert(len(dataframe.columns), 'y_Mean_Frequency_cop',value=float)
    dataframe.insert(len(dataframe.columns), 'y_Total_power_cop',value=float)

    
    return

cria_feature(bds)

print(bds)
```
